```python
#importing libs
import torch
import torch.nn as nn
```

```python
# create model class
# basically a single nural
class Model(nn.Module):
# "nn.module" is inherited here
  def __init__(self, num_features):
# constructor, num-features = no of rows
    super().__init__()
    # super keyword is used to invoke the constructor of nn.moduls class
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, features):

    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

```python
# create dataset
features = torch.rand(10,5)
```

```python
# create model
model = Model(features.shape[1])
```

```python
# call model for forward pass
# "model.forward(features)" this also works but pytorch recommends to code us like this as given below
model(features)
```

the above output is the y_prid of the nural network
