```python
#importing libs
import torch
import torch.nn as nn
```

```python
# create model class
# basically a single nural
class Model(nn.Module):
# "nn.module" is inherited here
  def __init__(self, num_features):
# constructor, num-features = no of rows
    super().__init__()
    # super keyword is used to invoke the constructor of nn.moduls class
    self.linear = nn.Linear(num_features,3)
      #first layer where there 5 features are given as input
    self.relu = nn.ReLU()
      # relu is used as bias
    self.linear2 = nn.Linear(3,1)
      # here there is three layers in hidden layer, and three inputs from 1st layer and 1 y_pred is given as output
    self.sigmoid = nn.Sigmoid()
      # sigmoid is used as bias

  def forward(self,features):
    out = self.linear(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)

    return out
```

```python
# create dataset
features = torch.rand(10,5)
```

```python
# create model
model = Model(features.shape[1])
```

```python
# call model for forward pass
# "model.forward(features)" this also works but pytorch recommends to code us like this as given below
model(features)
```

```python
# weights of first layer i.e
model.linear.weight
```

```python
# weights of the second layer
model.linear2.weight
```

```python
# ReLU bias of first layer
model.linear.bias
```

```python
# sigmoid bias of hidden layer
model.linear2.bias
```
