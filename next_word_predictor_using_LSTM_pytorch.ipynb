```python
#Tokenization
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
#tokenize
tokens = word_tokenize(document.lower())
```

```python
# built dictionary

vocab = {'<unk>':0}
# 'counter' is a builtin function that stores all the unique words in a dictionary and also tell us how many times does this word occured in the data
# this for loop adds every word unique words in vocab variable and also adds indexes to it
for token in Counter(tokens).keys():
  if token not in vocab:
    vocab[token] = len(vocab)
  # our vocablary is now completed
```

```python
vocab
```

```python
# extract sentences from data, basically divided the data into oneline sentences
input_sentences = document.split('\n')
```

```python
input_sentences
```

```python
# this particular function convert sentences into vectors
def text_to_indices(sentence, vocab):

  numerical_sentence = []

  for token in sentence:
    if token in vocab:
      numerical_sentence.append(vocab[token])
    else:
      numerical_sentence.append(vocab['<unk>'])

  return numerical_sentence
```

```python
# now embed these english sentences/words in numbers with resprct to the keys present in vocablary
input_numerical_sentences = []

for sentence in input_sentences:
  input_numerical_sentences.append(text_to_indices(word_tokenize(sentence.lower()), vocab))
```

```python
len(input_numerical_sentences)  # we have 78 sentences in total
```

```python
# this loop makes the supervied data from the given sentences in data, ones refer notes
training_sequence = []
for sentence in input_numerical_sentences:

  for i in range(1, len(sentence)):
    training_sequence.append(sentence[:i+1])
```

```python
# padding is needed here as the length of each vector is not equal so we have to bring all the vectors in same length
```

```python
# first lets find the max length of vector and apply the padding accordlingly such that all the vextors come in one lenght i.e the max lenght
len_list = []

for sequence in training_sequence:
  len_list.append(len(sequence))

max_len = max(len_list)
```

```python
max_len # now we have to bring all the vectors in the size of 62 by appling "pre padding" where zeros are added before the original vector
```

```python
# generiting zeros and storing zeros in the training sequences

padded_training_sequence = []

for sequence in training_sequence:
  padded_training_sequence.append([0]*(max(len_list) - len(sequence)) + sequence)
```

```python
# input and output data has been generated. x = input coulmn, y = output coulmn
x = padded_training_sequence[:,:-1]
y = padded_training_sequence[:,-1]
```

```python
# creating dataset class
class CustomDataset(Dataset):

  def __init__(self,x,y):
    super().__init__()
    self.x = x
    self.y = y

  def __len__(self):
    return self.x.shape[0]


  def __getitem__(self, idx):
    return self.x[idx], self.y[idx]
```

```python
dataset = CustomDataset(x,y) # making object of dataset class
```

```python
len(dataset) # to get the length of dataset
```

```python
dataset[0] # to retrive a particular tensor from dataset, retrived below is the first item from dataset
```

```python
# lets make a dataloader and its object
dataloader = DataLoader(dataset, batch_size=32,shuffle=True)
```

```python
# printed 32 items of each group
for input, output in dataloader:
  print(input,output)
```

```python
# the data is furtheer embedded in the form of (32,61,100) where 32 - batches, 61-words, 100dimentions of data
```

```python
from torch.nn.modules import Embedding
class LstmModel(nn.Module):
  def __init__(self,vocab_size,):
    super().__init__()
    self.embeddings = nn.Embedding(vocab_size,150)
    self.lstm = nn.LSTM(150,150,batch_first=True)
    self.fc = nn.Linear(150,vocab_size)

  def forward(self,x):
    embedded = self.embeddings(x)
    intermediate_hiddenstates, (final_hidden_state,final_cell_state) = self.lstm(embedded)
    output = self.fc(final_hidden_state.squeeze(0))
    return output
```

```python
model = LstmModel(len(vocab))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model.to(device)
```

```python
epochs = 50
learning_rate = 0.001

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)
```

```python
# training loop

for epochs in range(epochs):
  total_loss = 0

  for batch_x, batch_y in dataloader:
    batch_x = batch_x.to(device)
    batch_y = batch_y.to(device)
    optimizer.zero_grad()
    output = model(batch_x)
    loss = criterion(output,batch_y)
    loss.backward()
    optimizer.step()

    total_loss += loss.item()

  print(f"Epoch: {epochs+1}, Loss: {total_loss:.4f}")
```

```python
# prediction

def prediction(model, vocab, text):

  # tokenize
  tokenized_text = word_tokenize(text.lower())

  # text -> numerical indices
  numerical_text = text_to_indices(tokenized_text, vocab)

  # padding
  padded_text = torch.tensor([0] * (61 - len(numerical_text)) + numerical_text, dtype=torch.long).unsqueeze(0).to(device)

  # send to model
  output = model(padded_text)

  # predicted index
  value, index = torch.max(output, dim=1)

  # merge with text
  return text + " " + list(vocab.keys())[index]
```

```python
# Function to calculate accuracy
def calculate_accuracy(model, dataloader, device):
    model.eval()  # Set the model to evaluation mode
    correct = 0
    total = 0

    with torch.no_grad():  # No need to compute gradients
        for batch_x, batch_y in dataloader1:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)

            # Get model predictions
            outputs = model(batch_x)

            # Get the predicted word indices
            _, predicted = torch.max(outputs, dim=1)

            # Compare with actual labels
            correct += (predicted == batch_y).sum().item()
            total += batch_y.size(0)

    accuracy = correct / total * 100
    return accuracy

# Compute accuracy
accuracy = calculate_accuracy(model, dataloader, device)
print(f"Model Accuracy: {accuracy:.2f}%")
```
